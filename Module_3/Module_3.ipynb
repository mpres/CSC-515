```python
%%writefile Module_3.py
import cv2 as cv

#Step 1: extract image
face_img = cv.imread("front_face.jpg")

#Step 2:Draw red box left eye
cv.rectangle(face_img,(1205,825),(1375,905),(0,0,255),thickness=2)

#Step 3:Draw red box right eye
cv.rectangle(face_img,(1615,825),(1795,905),(0,0,255),thickness=2)

# Step 4. Draw circle around the face
cv.circle(face_img,(1500,850),225,(0,255,0),thickness=2)



#Step 5: show image with red rectangle
cv.imshow("Image with Red box", face_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
